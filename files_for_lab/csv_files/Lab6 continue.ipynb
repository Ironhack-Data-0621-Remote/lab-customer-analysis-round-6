```python
#One Hot/Label Encoding (categorical).
y = data1['income']
X = data1.drop(['income'], axis=1)
```

```python
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = object)
```

```python
cat_data = pd.get_dummies(X_cat, drop_first=True) 
cat_data[:20]
```

```python
encoder = OneHotEncoder(handle_unknown='error', drop='first')
encoder.fit(X_cat)

encoder = encoder.transform(X_cat).toarray()
encoder #.shape #i'm not sure
```

```python
encoder.shape
```

```python
#Concat DataFrames
x_normalized.shape, encoder.shape
```

```python
#Linear Regression
#Train-test split
X = np.concatenate([x_normalized, encoder], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Apply linear regression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
predictions  = model.predict(X_test)
predictions.shape
```

```python
predictions
```

```python
r2_score(y_test, predictions), mean_squared_error(y_test, predictions, squared=False), mean_squared_error(y_test, predictions)
```

```python
#Model Validation
#Get R2 from the model
r2 = r2_score(y_test, predictions)
print(r2)
```

```python
#Get MSE from the model
mse = mean_squared_error(y_test, predictions)
print(mse)
```

```python
#Get RMSE from the model
rmse = math.sqrt(mse)
print(rmse)
```

```python
#Get MAE from the model
n = len(X_test)
p = X_test.shape[1]
adj_r2 = 1-((1-r2)*(n-1)/(n-p-1))
print(adj_r2)
```
